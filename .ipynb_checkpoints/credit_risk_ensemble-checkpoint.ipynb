## Read the CSV and Perform Basic Data Cleaning

```python
# Load the data
file_path = Path("Resources/LoanStats_2019Q1.csv")
df = pd.read_csv(file_path)

# Preview the data
df.head()
```

```python
df.dtypes
```

```python
df["home_ownership"].count
```

```python
from sklearn.preprocessing import LabelEncoder, StandardScaler
# Creating an instance of label encoder
label_encoder = LabelEncoder()
```

```python
# Fitting the label encoder
label_encoder.fit(df["home_ownership"])
```

```python
# List the classes identified by the label encoder
list(label_encoder.classes_)
```

```python
# Encode the months as an integer
df["home_ownership_le"] = label_encoder.transform(df["home_ownership"])
df.tail()
```

```python
label_encoder.fit(df["initial_list_status"])
```

```python
list(label_encoder.classes_)
```

```python
# Encode the months as an integer
df["initial_list_status_le"] = label_encoder.transform(df["initial_list_status"])
df.tail()
```

```python
label_encoder.fit(df["issue_d"])
```

```python
list(label_encoder.classes_)
```

```python
# Encode the months as an integer
df["issue_d_le"] = label_encoder.transform(df["issue_d"])
df.tail()
```

```python
# howeowner dictionary
home_ownership_num = {
    "ANY": 0,
    "MORTGAGE": 1,
    "OWN": 2,
    "RENT": 3,
}
```

```python
# howeowner' names encoded using the dictionary values
df["home_ownership_num"] = df["home_ownership"].apply(lambda x: home_ownership_num[x])
df.head()
```

```python
# Dropping month and month_le columns
df.drop(["home_ownership", "home_ownership_le", "initial_list_status", "issue_d"], axis=1, inplace=True)
df.head(20)
```

```python
# Binary encoding using Pandas (single column)
LoanStats_encoded = pd.get_dummies(df)
LoanStats_encoded.head()
```

```python
# Saving the encoded dataset
file_path = Path("Resources/loanStats_encoded.csv")
LoanStats_encoded.to_csv(file_path, index=False)
```

## Split the Data into Training and Testing

```python
# Create our features
X = df.drop(columns="loan_status")
X = pd.get_dummies(X)
# Create our target
y = df["loan_status"]
```

```python
X.describe()
```

```python
# Check the balance of our target values
y.value_counts()
```

```python
X.dtypes
```

```python
# Split the X and y into X_train, X_test, y_train, y_test
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, 
                                                    y, 
                                                    random_state=1, 
                                                    stratify=y)
X_train.shape
```

## Data Pre-Processing

Scale the training and testing data using the `StandardScaler` from `sklearn`. Remember that when scaling the data, you only scale the features data (`X_train` and `X_testing`).

```python
# Create the StandardScaler instance
from sklearn.preprocessing import LabelEncoder, StandardScaler
data_scaler = StandardScaler()
```

```python
# Fit the Standard Scaler with the training data
# When fitting scaling functions, only train on the training dataset
X_scaler = data_scaler.fit(X_train)
```

```python
# Scale the training and testing data
X_train_scaler= X_scaler.transform(X_train)
X_test_scaler = X_scaler.transform(X_test)
```

### Balanced Random Forest Classifier

```python
# Resample the training data with the BalancedRandomForestClassifier
from imblearn.ensemble import BalancedRandomForestClassifier
brf = BalancedRandomForestClassifier(n_estimators=100, random_state=1)
brf.fit(X_train, y_train)
```

```python
# Calculated the balanced accuracy score
y_pred = brf.predict(X_test_scaler)
balanced_accuracy_score(y_test,y_pred)
```

```python
# Display the confusion matrix
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, y_pred)
```

```python
# Print the imbalanced classification report
print(classification_report_imbalanced(y_test, y_pred))
```
